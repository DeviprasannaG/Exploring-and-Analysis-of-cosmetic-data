```python
import pandas as pd
cosmetic_df= pd.read_csv(r"C:\Users\gdevi\Desktop\final_dataset.csv", sep=",",encoding='latin-1')
cosmetic_df
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
# Combine all the ingredients into a single string
all_ingredients = " ".join(cosmetic_df["ingredients"].str.lower().dropna())
# Create a word cloud from the combined ingredients string
wordcloud = WordCloud(width=800, height=800, background_color="white").generate(all_ingredients)
# Display the word cloud
plt.figure(figsize=(5,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
